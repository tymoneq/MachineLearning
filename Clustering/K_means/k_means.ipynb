```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
dataSet = pd.read_csv("Mall_Customers.csv")

x = dataSet.iloc[:, 2:].values
```

```python
wcss = []

for i in range (1,11):
    kmeans = KMeans(n_clusters=i,init="k-means++", random_state= 2137)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
    

print(wcss)
```

```python
plt.plot(range (1,11), wcss)
```

```python
kmeans = KMeans(n_clusters=5,init="k-means++", random_state= 2137)
types = kmeans.fit_predict(x)
```
